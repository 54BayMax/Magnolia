```python
import numpy as np
import gennoisy
import matplotlib.pylab as plt
import pandas as pd

%matplotlib inline
```

```python
def sigmoid(X):
    return 1./(1 + np.exp(-X))

def dsigmoid(X):
    return sigmoid(X)*(1 - sigmoid(X))
```

```python
X,S,A,N = gennoisy.gennoisy()
```

```python
plt.plot(X)
```

```python
# W = np.random.rand(*A.shape)

def likelihood(W,X):
    return sigmoid(  W.dot(X.T)  ).prod(axis=0) * np.linalg.det(W)

def update(W, X, alpha=0.001):
    return alpha*(1 - 2*sigmoid(W.dot(X.T))).dot(X) + np.linalg.inv( W.T )

losses = np.zeros(100)
for j in range(100):
    for i in range(1000):
        W = W + update(W, X)
    losses[j]=np.log( likelihood(W,X) ).sum()

print(losses)
```

```python
import numpy as np
import gennoisy
from ica import ica
import matplotlib.pylab as plt
%matplotlib inline
```

```python
X,S,A,N = gennoisy.gennoisy()
plt.plot(X)
plt.title('Mixed Signals (3 channels)')
plt.figure()
plt.plot(S)
plt.title('Original Signals')
```

```python
ivecs,m = ica(X)
rX = (X-m).dot(ivecs).dot( np.linalg.inv(ivecs.T.dot(ivecs)))
plt.plot(rX)
plt.title('Rotated mixed')
```

```python
import numpy as np
import matplotlib.pylab as plt
%matplotlib inline
from scipy.signal import sawtooth
```

### Generate Noisy Mixed Signals

```python
def gennoisy(S=None, A=None, N=None):

    # A is the channel
    A = np.random.rand(3,3);

    # Time with sampling frequency 200 Hz
    fs = 1.0/200;
    t = np.zeros(2000);
    t[:] = np.arange(0,10,fs);

    # Signal
    s1 = np.sin(2 * t)  # Signal 1 : sinusoidal signal
    s2 = np.sign(np.sin(3 * t))  # Signal 2 : square signal
    s3 = sawtooth(2 * np.pi * t)  # Signal 3: saw tooth signal
    S = np.array([s1,s2,s3]).T;

    # Generate the noise
    N = .05*np.random.randn(*S.shape);

    # Final microphone has elements of each signal
    X = S.dot(A)+N;

    return X, S, A, N
```

```python
X,S,A,N = gennoisy()
plt.subplot(211); plt.plot(X); plt.title('Mixed Signals (3 channels)')
plt.figure()
plt.subplot(212); plt.plot(S); plt.title('Original Signals')
```

```python
from sklearn.decomposition import FastICA
def ica(X, sigdim=0):
    if sigdim==0:
        Xm = X
    else:
        Xm = X.T
    ica = FastICA(n_components=3)
    S_ = ica.fit_transform(Xm)
    A_ = ica.mixing_
    m_ = ica.mean_
    return A_, m_
```

```python
rmat, m = ica(X)
Sn = (X - m).dot(rmat).dot( np.linalg.inv( rmat.T.dot(rmat) ) )
plt.plot(Sn)
plt.title('Recovered Signals')
```
